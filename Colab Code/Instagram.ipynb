# ***MATPLOTLIB , SEABOARN***




```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="rainbow_r")
```

Darker shades of color (close to red in the "rainbow_r" colormap) represent strong positive correlations (values close to +1). These indicate pairs of features that tend to increase together. For example, if "Feature A" and "Feature B" have a correlation coefficient close to +1, it suggests that higher values of "Feature A" are associated with higher values of "Feature B".

```python
plt.scatter(df["Followers"], df["Month"])
plt.xlabel("Followers")
plt.ylabel("Month")
plt.title("Followers vs Month")
plt.show()
```

If the points on the scatter plot show an upward or downward trend, this could indicate a general increase or decrease in the number of followers over the months.
If the points are scattered without any discernible trend, it suggests that there is no clear relationship between the number of followers and the months.


```python
top_unique_followed_person = df.nlargest(10, 'Followers').drop_duplicates(subset=['Username'])
top_unique_followed_person
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Username', y='Followers', data=top_unique_followed_person, palette='viridis')
plt.xlabel('Username')
plt.ylabel('Followers')
plt.title('Top 5 Influencers by Number of Followers')
plt.show()
```

The bar plot created by the provided code visualizes the number of followers for the top 4 influencers.

```python
copy_df  = df.copy()
```

```python
max = copy_df['Followers'].max()
min = copy_df['Followers'].min()
mean = copy_df['Followers'].mean()
median = copy_df['Followers'].median()
std_dev =copy_df['Followers'].std()
var = copy_df['Followers'].var()
skew = copy_df['Followers'].skew()
kurt = copy_df['Followers'].kurt()
print(f'max: {max}')
print(f'min: {min}')
print(f"mean: {mean}")
print(f'median: {median}')
print(f'standard deviation: {std_dev}')
print(f'variance: {var}')
print(f'skewness: {skew}')
print(f'kurtosis: {kurt}')
```

```python
max_life = 2000
data_frame = pd.DataFrame({
    "mean":[mean for i in range(max_life)],
    "median":[median for i in range(max_life)],
    "std_dev":[std_dev for i in range(max_life)],
    "variance":[var for i in range(max_life)],
    "skewness":[skew for i in range(max_life)],
    "kurtosis":[kurt for i in range(max_life)]
})
data_frame
```

```python
plt.figure(figsize = (15,7))

plt.subplot(1,3,1)
data_frame["variance"].plot()
plt.legend(["variance"])

plt.subplot(1,3,2)
data_frame["mean"].plot()
data_frame["median"].plot()
data_frame["std_dev"].plot()
plt.legend(["mean", "median", "std_dev"])

plt.subplot(1,3,3)
data_frame["skewness"].plot()
data_frame["kurtosis"].plot()
plt.legend(["skewness", "kurtosis"])
plt.show()
```
