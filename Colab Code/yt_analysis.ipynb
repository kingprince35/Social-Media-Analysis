```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yt_june = pd.read_excel('Yt_filter_data.xlsx')
yt_nov = pd.read_excel('Yt_filter_data.xlsx','youtube -nov 2022')
yt_dec = pd.read_excel('Yt_filter_data.xlsx','Youtube_Filter_data')
df = pd.concat([yt_june,yt_nov,yt_dec],ignore_index=True)
df
```

Replace NA to 0 values


```python
df["Likes"] = df["Likes"].replace("N/A'" , '0')
df
```

```python
def convert(column):
    if 'M' in column:
        return float(column.replace('M', '')) * 1000000
    elif 'K' in column:
        return float(column.replace('K', '')) * 1000
    else:
        return float(column)
```

```python
df.info()
```

```python
df["Followers"] = df["Followers"].apply(convert)
df["Views"] = df["Views"].apply(convert)
df["Likes"] = df["Likes"].apply(convert)
df["Comments"] = df["Comments"].apply(convert)
df
```

# Hypothesis

```python
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
```

```python
ho = 0 #Views do not affect likes.
ha = 1 #Views do affect likes.

_ , p = ttest_rel(df["Likes"] , df["Views"])

print("p values =" , p)
alpha = 0.05
if(p < alpha):
  print("Reject Null Hypothesis (H0) hence,")
  print(ha)
else:
  print("Accept Null Hypothesis hence,")
  print(ho)
```
