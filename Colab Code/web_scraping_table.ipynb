```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://targetinternet.com/resources/fastest-growing-social-media-networks-2022-23/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
soup

# Find the table that contains the list of most-followed Instagram accounts
table = soup.find('table')

# Check if the table was found
if table:
    # Initialize a list to store the table data
    table_data = []

    # Loop through the rows of the table
    for row in table.find_all('tr'):
        # Extract the columns
        columns = row.find_all('td')
        # If there are columns in the row, extract their text content
        if columns:
            row_data = [col.get_text(strip=True) for col in columns]
            table_data.append(row_data)

    # Print the extracted table data
    for row in table_data:
        print(row)
else:
    print("Table not found.")
var  = pd.DataFrame(table_data)
var.to_csv('fastest_growing_netwrok_2022-23.csv')
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_most-followed_Instagram_accounts"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table that contains the list of most-followed Instagram accounts
table = soup.find('table', {'class': 'wikitable'})

# Check if the table was found
if table:
    # Initialize a list to store the table data
    table_data = []

    # Loop through the rows of the table
    for row in table.find_all('tr'):
        # Extract the columns
        columns = row.find_all('td')
        # If there are columns in the row, extract their text content
        if columns:
            row_data = [col.get_text(strip=True) for col in columns]
            table_data.append(row_data)

    # Print the extracted table data
    for row in table_data:
        print(row)
else:
    print("Table not found.")
```
